```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/사고 피해규모 정보_2021년.csv')
```

```python
#컬럼 이름 바꾸기

df=df.rename(columns={'TOXICITY_DIST':'독성거리'})
df=df.rename(columns={'ACDNT_TYPE':'사고유형'})
df=df.rename(columns={'DMGE_TYPE':'피해유형'})
df=df.rename(columns={'PIPE_ACDNT_DMGE_ARA':'배관사고 피해면적'})
df=df.rename(columns={'CO_INFO':'회사정보'})
df=df.rename(columns={'RDA_DIST':'방사거리'})
df=df.rename(columns={'EXPS_DIST':'폭발거리'})
df=df.rename(columns={'PIPE_DIA':'배관지름'})
df=df.rename(columns={'PIPE_LT':'배관길이'})
df=df.rename(columns={'PIPE_MART_CODE':'배관재료'})
df=df.rename(columns={'PIPE_OPRT_TP':'배관 운전온도'})
df=df.rename(columns={'PIPE_PRESSURE':'배관압력'})
df=df.rename(columns={'PIPE_FLRT':'배관유량'})
df=df.rename(columns={'PIPE_INSTL_YEAR':'배관설치연도'})
df=df.rename(columns={'PIPE_NO':'배관번호'})
```

```python
#필요없는 컬럼 삭제하기

df=df.drop(['ACDNT_DMGE_SCALE_UQNUM'],axis=1)
```

```python
df
```

```python
#사고유형 별 칼럼 평균

mean_type=df.groupby('사고유형').mean()
mean_type
```

```python
#회사정보 별 칼럼 평균

mean_coinfo=df.groupby('회사정보').mean()
mean_coinfo
```

```python
#배관번호 별 칼럼 평균

mean_pipeno=df.groupby('배관번호').mean()
mean_pipeno
```

```python
plt.figure(figsize=(8,5))
plt.title('사고유형 별 칼럼 평균')

plt.show()
```
